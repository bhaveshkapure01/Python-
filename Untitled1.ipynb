```python
n1=int(input())
n2=int(input())

# Find the maximum of the two numbers
max_val = max(n1, n2)

# Initialize lcm with the maximum value
lcm = max_val

# Loop until LCM is found
while lcm % n1 != 0 or lcm % n2 != 0:
  lcm += max_val # Increment lcm by max_val to find the next multiple

print(lcm)
```
