```python
import numpy as np
a = np.arange(2, 26, 2).reshape(3, 4)
print(a)
```

```python
a[:2]          #it print the rows but only 2 cols
a[2:]          #it print after exclude the 2 rows then print only single cols
a[:,::2]       #it print the only  2 elements
a[:-1]         #prints the only 2 cols, exclude the last row
a[:,1:3]       #it prints all rows and columns from index 1 to 2
```
